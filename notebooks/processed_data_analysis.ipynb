```python
df['date_mscan_yyyymm'] = df['date_mscan'].dt.to_period('M')
# df.reset_index()
```

```python
dfg = df.groupby(
    ["index_right", "date_mscan_yyyymm"],
    #as_index=False,
).agg(
    count_obs=pd.NamedAgg(column='id', aggfunc='count'),
    temp_bb_mean=pd.NamedAgg(column='temp_bb', aggfunc='mean'),
    temp_bb_sum=pd.NamedAgg(column='temp_bb', aggfunc='sum'),
)

dfg = dfg.reset_index()

dfg
```

```python
eia = pd.read_csv(
    '/Users/danielmsheehan/general_assembly/github/projects/project_6/capstone/data/input/eia/all_eia_data.csv',
    parse_dates=["month"]
)

eia['month_yyyymm'] = eia['month'].dt.to_period('M')

eia = eia[['region', 'month_yyyymm', 'oil_bbl_d_total_production', 'rig_count']]

eia = pd.merge(
    eia, 
    dfg, 
    left_on=['region', 'month_yyyymm'],
    right_on=['index_right', 'date_mscan_yyyymm'],
    how='outer',
)
```

```python
eia_permian = eia[eia['region'] == 'Permian Region']
eia_permian.set_index('month_yyyymm', inplace=True)
```
